```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import ast
```

```python
df=pd.read_csv("./animes.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['score'].fillna(np.mean(df['score']),inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['synopsis']=df['synopsis'].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
df.tail()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['synopsis'][1]
```

```python
df.drop(df[df['ranked']>1000].index,inplace=True,axis=0)
```

```python
df.tail()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
c_vec=cv.fit_transform(df['synopsis']).toarray()
```

```python
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
def stemm(text):
    l=[]
    for i in text.split():
        l.append(ps.stem(i))
    return " ".join(l)
```

```python
df['synopsis']=df['synopsis'].apply(stemm)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(c_vec)
```

```python
def getIndex(anime):
    anime_index=df[df['title']==anime].index[0]
    similar=similarity[anime_index]
    anime_list=sorted(list(enumerate(similar)),reverse=True,key=lambda x:x[1])[1:6]
    for i in anime_list:
        print(df.iloc[i[0]].title)
```
